```python
import pandas as pd
```

## load the JSON file that you created in the previous exercise into a dataframe

```python
file = pd.read_json("../data/auto.json", orient='records').set_index('CarNumber')
file
```

## make the following selects

```python
file[file['Fines'] > 2100]
```

```python
file[(file['Fines'] > 2100) & (file['Refund'] == 2)]
```

```python
file[(file['Model'] == 'Focus') | (file['Model'] == 'Corolla')]
```

```python
file[file.index.isin(['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS'])]
```

## make the aggregations with the make and the model

```python
file.groupby(by=['Make']).agg({'Fines': ['median']})
```

```python
file.groupby(by=['Make', 'Model']).agg({'Fines': ['median']})
```

```python
file.groupby(by=['Make', 'Model']).agg({'Fines': ['count']})
```

```python
file.groupby(by=['Make', 'Model']).agg({'Fines': ['min', 'max']})
```

```python
file.groupby(by=['Make', 'Model']).agg({'Fines': ['std']})
```

```python
c = file.groupby(by=['CarNumber']).agg({'Fines': ['count']}).sort_values(('Fines', 'count'), ascending=False)
c
```

```python
file.loc[c.index[0]]
```

```python
c = file.groupby(by=['CarNumber']).agg({'Fines': ['sum']}).sort_values(('Fines', 'sum'), ascending=False)
c
```

```python
file.loc[c.index[0]]
```
